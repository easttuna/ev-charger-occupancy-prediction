```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('classic') 
df = pd.concat([pd.read_csv(f'../metric_table/metrics_testfrac-{round(frac, 1)}_predstep-1.csv') for frac in np.arange(0.1, 1, 0.1)])
model_name_encoder = {name:name for name in df.model}
model_name_encoder.update({'MultiSeqHybrid':'MultiSeqFusion', 'MultiSeqHybridUmap':'MultiSeqFusion(+umap embedding)'})
df.model = df.model.map(model_name_encoder)

ref_df = pd.read_csv('../metric_table/timewise_metrics_testfrac-0.2_predstep-1.csv')
```

```python
import seaborn as sns

cand_models = ['MultiSeqFusion', 'MultiSeqFusion(+umap embedding)']
# cand_models = ['MultiSeqBase', 'MultiSeqHybridUmap']
# cand_models = ['MultiSeqHybrid', 'MultiSeqHybridUmap', 'MultiSeqHybridUmapEarlyGating']

fig, ax = plt.subplots(figsize=(16,8))
data = df[df.model.isin(cand_models)].pivot(index='test_frac', columns='model', values='f1_weighted')
data['max score*'] = 0.852210

sns.lineplot(data=data, ax=ax, lw=2.5)
ax.grid(True)
ax.set_title('F1 Score by Test Data fraction (pred step 1)', fontsize=20)
ax.set_xlabel('test fraction', fontsize=15)
ax.set_ylabel('F1 score (weighted)', fontsize=15)
# ax.legend(loc='lower left', prop={'size':20})

leg = plt.legend(loc='lower left')
# get the lines and texts inside legend box
leg_lines = leg.get_lines()
leg_texts = leg.get_texts()
# bulk-set the properties of all lines and texts
plt.setp(leg_lines, linewidth=4)
plt.setp(leg_texts, fontsize='x-large')
plt.show()
```
